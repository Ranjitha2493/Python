<h2>Question One </h2>
 
Write a program that accepts :
-  the names and prices of all the fruits in inventory, and 
-  the names and quantities of each fruit that was purchased 

Your program should print out the total cost per fruit that was purchased and also the final amount owed by the customer.

<h3>Sample Interaction</h3>



````python
n = int(input("Enter the number of fruits in your inventory: "))
i = 1
d = {}
while i <=n:
    f = input("Enter the name of the fruit: ")
    f = f.lower()
    p = float(input("Enter the price of the fruit: "))
    d[f] = p
    i+=1
#create the pattern as per the sample output
var = '*' * 13
print(var)

n1 = int(input("Enter the number of fruits purchased: "))
print(var)

i=1
sum = 0
while i<=n1:
    f = input("Enter the name of fruit purchased: ")
    f = f.lower()
    p = int(input("Enter the quantity purchased: "))
    print('\nThe cost of {0} {1} is ${2:0.2f}\n'.format(p,f,d[f]*p))
    sum += d[f]*p
    i+=1
print(var)
print('The total amount spent was ${0:0.2f}'.format(sum))
````

````python
import csv
st_name = []
st_score = []
with open('MIS6382 Fall2021-HW3Q2IInput.csv', newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    for row in csvreader:
        st_name.append(row[0].strip())
        st_score.append(int(row[1]))

basket = input("Enter the list of students who play basketball. Separate the names with ',' : \n")
basket = basket.split(',')
basket = [x.strip(' ') for x in basket] #remove extra space
basket = set(basket)

volley = input("\nEnter the list of students who play volleyball. Separate the names with ',' : \n")
volley = volley.split(',')
volley = [x.strip(' ') for x in volley] #remove extra space
volley = set(volley)

only_basket = basket.difference(volley)
only_volley = volley.difference(basket)
both_game = basket.intersection(volley)
print("\nStudents who only play Basketball",only_basket)
print("\nStudents who only play Volleyball",only_volley)
print("\nStudents who play both games: ",both_game)

#student who dont play basketball nor volleyball
nobv = []
for i in range(0,len(st_name)):
    if(st_name[i] not in basket and st_name[i] not in volley):
        nobv.append(st_name[i])
nobv = set(nobv)

print("\nStudents who play neither game: ",nobv)

print("\nThe names of all students along with their adjusted scores is:")

for i in only_basket:
    if (st_score[(st_name.index(i))]>=92):
        st_score[(st_name.index(i))] = 100
    else:
        st_score[(st_name.index(i))] += 8
    
for i in only_volley:
    if(st_score[(st_name.index(i))] >=95):
        st_score[(st_name.index(i))] = 100
    else:
        st_score[(st_name.index(i))] += 5
    
for i in both_game:
    if (st_score[(st_name.index(i))] >= 90):
        st_score[(st_name.index(i))] = 100
    elif(st_score[(st_name.index(i))] >=80):
        st_score[(st_name.index(i))] += 10
    elif(st_score[(st_name.index(i))] <=5):
        st_score[(st_name.index(i))] = 0
    else:
        st_score[(st_name.index(i))] -= 5

l = 0
for i in st_name:
    if(len(i) > l):
        l = len(i)
#print in output format 15 is length of letter of name      
for i in st_name:
    l= len(i)
    print(i,(15-l)*" ",":",st_score[(st_name.index(i))])
    
````

<h2>Question Four </h2>

Write a Python program that accepts a list of passwords (separated by commas) from the user. For each password in the list, the program should call a function to verify whether or not the password is valid.  Valid passwords should meet the following rules:

-  a.	At least 8 characters and no more than 15 characters
-  b.   Must begin with two lower case alphabets   
-  c.	At least one occurance of two consecutive uppercase letters
-  d.   At least one occurance of a two digit number
-  e.	Exactly one of the following special characters (#, %, @, &)	


<h3>Sample Interaction</h3>



````python
import re
def check_password(s):
    if(len(s)<8 or len(s)>15):
        print(f'\n{s} does not match password rules')
        print("Password must be between 8 and 15 characters long!")
    elif not (re.search('^[a-z][a-z]',s)):
        print(f'\n{s} does not match password rules')
        print("Password must start with two lower case letters!")
    elif not (re.search('[A-Z][A-Z]',s)):
        print(f'\n{s} does not match password rules')
        print("Password must contain atleast one occurance of two consecutive uppercase letters!")
    elif not(re.search('\d\d',s)):
        print(f'\n{s} does not match password rules')
        print("Password must contain atleast one occurance of two consecutive digits!")
    elif (len(re.findall('#|%|@|&',s))>1):
        print(f'\n{s} does not match password rules')
        print("Password must contain exactly one occurance of the special characters #%@& !")
    else:
        print(f'\n{s} matches password rules')
        

password = input("Enter the passwords separated by commas:\n")
password = password.split(',')
for i in range(0,len(password)):
    password[i] = password[i].strip(" ")
    check_password(password[i])
````
