<h2>Question One </h2>

Write a program to ask the user to enter the speed of her car.  Then the program should display the distance 
travelled by the car in 

__7 hours, 10 hours and 14 hours__.

Assuming there are no delays or accidents, the distance that a car travels down the interstate can be calculated with the following formula 

__Distance = Speed x Time__

<h3>Sample Interaction</h3>



```python
speed = int(input("Enter the speed of your vehicle: ")) # assuming user will enter int value only

if(speed < 0):
    print("Enter speed greater than 0, exiting the program")
else:
    print() #adding an extra line as in sample output
    print(f'The distance the car will travel in 7 hours is {speed * 7} miles')
    print(f'The distance the car will travel in 10 hours is {speed * 10} miles')
    print(f'The distance the car will travel in 14 hours is {speed * 14} miles')
```

<h2>Question Two </h2>

Write a program that asks the user to enter a positive integer, n, that is less than 10. If the user enters a 
negative number or a number greater than 9, the program should display a message and stop.  
Otherwise the program should print the the result from evaluating (nn * nnn-n)/nn. Your result should be rounded to three decimal points.
For example, if the user enters 8, the program should print 887.909.

<h3>Sample Interaction</h3>



<h3>Sample Interaction</h3>



```python
n = int(input("Enter a positive integer less than 10: "))

if(n>9 or n < 0):
    print("Incorrect input. Exiting the program")
else:
    nn = n*10 + n
    nnn = n*100 + nn
    result = (nn * nnn - n)/nn
    print(f'The result is: {round(result,3)}')
```

```python
weight = float(input("Enter the weight of the package: "))

if(weight < 0):
    print("Incorrect weight, exiting the program")
elif(weight <= 1.5):
    print('The cost of shipping the package is: ${0:0.2f}'.format(weight*2.40))    
elif(weight <= 5 and weight > 1.5):
    print('The cost of shipping the package is: ${0:0.2f}'.format(weight*3.20))
elif(weight >5 and weight <= 10):
    print('The cost of shipping the package is: ${0:0.2f}'.format(weight*3.70))
else:
    print('The cost of shipping the package is: ${0:0.2f}'.format(weight*4.20))
```
