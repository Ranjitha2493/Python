<h2>Question Two </h2>

Write a program that accepts a string as input from the user and then does the following:
1. counts the number of numeric characters in the string and prints 
2. removes these charaters from the string
3. creates a new string without these characters
4. prints out the count of numeric characters, and the new string

__Hint:__ To find out if a character is numeric use the `isnumeric()` method.

<h3>Sample Interaction</h3>



```python
string = input("Enter a string: ")
new_string = ''
count = 0

for i in string:
    if i.isnumeric():
        count+= 1
        continue
        
    else:
        new_string += i

print(f'There were {count} numeric characters in the input string')
print(f'The ouput string without numeric charcaters is: {new_string}')
```
