```python
class Vehicle:
    
#constructor
    def __init__(self, make, model, year_manufacture, mileage):
        self.__make = make
        self.__model = model
        self.__year_manufacture = year_manufacture
        self.__mileage = mileage 

#accessor methods
    def get_make(self):
        return self.__make

    def get_model(self):
        return self.__model

    def get_mileage(self):
        return self.__mileage

    def get_year_manufacture(self):
        return self.__year_manufacture
 
 #mutator methods
    def set_make(self, mk):
        self.__make = mk

    def set_model(self, md):
        self.__model = md

    def set_mileage(self, m):
        self.__mileage = m

    def set_year_manufacture(self, ym):
        self.__year_manufacture = ym

#redefined __str__ method
    def __str__(self):
        return 'Make: ' + self.__make + '; Model: ' + self.__model + '; Year of Manufacture: ' + str(self.__year_manufacture) + '; Mileage ' + str(self.__mileage)
 
```

```python
class Employee:
    
# constructor
    def __init__(self, name, address, v):
        self.__name = name
        self.__address = address
        self.__vehicle = v
        
# accessor methods
    def get_name(self):
        return self.__name

    def get_address(self):
        return self.__address

    def get_vehicle(self):
        return self.__vehicle
    
    def calc_compensation(self):
        pass

# mutator methods
    def set_name(self, n):
        self.__name = n

    def set_address(self, ad):
        self.__address = ad

    def set_vehicle(self, v):
        self.__vehicle = v
    
# redefined __str__ method
    def __str__(self):
        emp= 'Employee Name: ' + self.__name + '; Employee Address: ' + self.__address 
        emp+='\n'+self.__vehicle.__str__() +'\n'
        return emp
```

```python
class FullTimeEmployee(Employee):

#constructor
    def __init__(self, name, address, v, sal): 
        super().__init__(name, address, v)
        self.__salary=sal
        self.emptypeflag = "F" #to identify the type of employee
    
#accessor method
    def get_salary(self):
        return self.__salary
    
#mutator method
    def set_salary(self, sal):
        self.__salary = sal  
        
#compensation calculation         
    def calc_compensation(self):
        if self.__salary <= 40000:
            tx=(16/100) * self.__salary
        elif 40000<self.__salary<=75000:
            tx=(16/100)*40000+(26/100)*(self.__salary-40000)
        elif self.__salary>75000:
            tx=(16/100)*40000+(26/100)*(75000-40000)+(33/100)*(self.__salary-75000)
        
        comp=self.__salary - tx
        return comp
    
#redefine __str__ method
    def __str__(self):
        full_detail=super().__str__()
        full_detail +="Salary: {:0.2f}".format(self.__salary) + '\n'
        return full_detail
```

```python
class HourlyEmployee(Employee):
    
#constructor
    def __init__(self,name, address, v, hr_wrk, hr_rt): 
        super().__init__(name, address, v)
        self.__hr_worked=hr_wrk
        self.__hr_rate=hr_rt
        self.emptypeflag = "P" #to identify the type of employee
    
#accessor methods
    def get_hr_worked(self):
        return self.__hr_worked
    
    def get_hr_rt(self):
        return self.__hr_rate
    
#mutator methods
    def set_hr_worked(self,hr_wrk):
        self.__hr_worked=hr_wrk
    
    def set_hr_rt(self,hr_rt):
        self.__hr_rate=hr_rt
        
#compensation calculation  
    def calc_compensation(self):
        if self.__hr_worked<=40:
            comp=self.__hr_worked*self.__hr_rate
        else:
            comp=40*self.__hr_rate+(self.__hr_worked-40)*self.__hr_rate*1.7
        return comp
    
#redefine __str__ method
    def __str__(self):
        hourly_detail=super().__str__()
        hourly_detail +="Hours Worked: {}; Hourly rate: {:0.0f}".format(self.__hr_worked,self.__hr_rate) + '\n'
        return hourly_detail
```

```python
class Consultant(Employee):

#constructor
    def __init__(self,name, address, v, hr_wrk, project): 
        super().__init__(name, address, v)
        self.__hr_worked=hr_wrk
        self.__proj=project
        self.emptypeflag = "C" #to identify the type of employee
    
#accessor methods
    def get_hr_worked(self):
        return self.__hr_worked
    
    def get_project_type(self):
        return self.__proj
    
#mutator methods
    def set_hr_worked(self,hr_wrk):
        self.__hr_worked=hr_wrk
    
    def set_project_type(self,project):
        self.__proj=project
        
#compensation calculation    
    def calc_compensation(self):
        if self.__proj == 1:
            rate=55
        elif self.__proj == 2:
            rate=70
        elif self.__proj == 3:
            rate=85
        
        comp=self.__hr_worked*rate
        return comp
    
#redefine __str__ method
    def __str__(self):
        consultant_detail=super().__str__()
        consultant_detail +="Hours Worked: {}; Project Type: {}".format(self.__hr_worked,self.__proj) +'\n'
        return consultant_detail      
```

```python
#to open a file and add data to the file
def read_file():
    my_lst = [] # creating empty list
    try:
        fb = open('emp.dat', 'rb') 
    except FileNotFoundError:
        return my_lst 

    else: 
        try:
            while (True): 
                my_lst.append(pickle.load(fb))
        except EOFError: 
            fb.close()
            return my_lst
```

```python
#print the employee details

def get_employee(emp_lst):
    count = 1
    for emp in emp_lst:
        if emp.emptypeflag == "F":
            if(count == 1):
                print("\nDetails of Full Time Employee are:")
                count = 0
            det = emp.__str__()
            print(det)
 
    count = 1
    for emp in emp_lst:
        if emp.emptypeflag == "C":
            if count == 1:
                print("\nDetails of Consultant are:")
                count = 0
            det = emp.__str__()
            print(det)
            
    count = 1
    for emp in emp_lst:
        if emp.emptypeflag == "P":
            if count == 1 :
                print("\nDetails of Hourly Employee are:")
                count = 0
            det = emp.__str__()
            print(det)
```

```python
#print the name and compensation of all the employees

def get_emp_compensation(emp_lst):
    print("\nEmployee name and Compensation of all Employees")
    print("=================================================")
    for emp in emp_lst:
        n = emp.get_name()
        comp = emp.calc_compensation()
        print("{0}'s is ${1:0.2f}".format(n,comp))
    print()   
```

```python
#print the employee name and the vehicle details if the mileage is greater than 78000

def get_emp_name_vehicle(emp_lst):
    print("\nEmployee Name and Vehicle Data For High Mileage Vehicles")
    print("============================================================")
    for emp in emp_lst:
        mil = emp.get_vehicle().get_mileage()
        if mil > 78000:
            n = emp.get_name()
            v = emp.get_vehicle()
            print(f'Employee name is: {n} {v}')
        else:
            continue
```
